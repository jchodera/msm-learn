```python
from numpy.linalg import det
BC = lambda X,Y: det(X.T.dot(Y))/ np.sqrt(det(X.T.dot(X)) * det(Y.T.dot(Y)))
```

```python
%timeit BC(pos[0],pos[-1000])
```

```python
from time import time

t = time()

n = len(pos)
bc_mat = np.zeros((n,n))

for i in range(n):
    for j in range(i):
        bc_mat[i,j] = BC(pos[i],pos[j])
    if i%10==0:
        print(i)
bc_mat = bc_mat + bc_mat.T
print(time()-t)
```

```python
pl.rc('font', family='serif')
pl.imshow(bc_mat,interpolation='none')
pl.colorbar()
pl.title('Pairwise Binet-Cauchy kernel: Alanine Dipeptide')
pl.savefig('../projects/rmsd/figures/ala-dipeptide-bc-10k.jpg',dpi=600)
```

```python
# test how well RMSD obeys the triangle inequality
# d(A,C) <= d(A,B) + d(B,C)
```

```python
triangle = np.zeros((100,100,100))

for i in range(100):
    for j in range(i,100):
        for k in range(j,100):
            triangle[i,j,k] = rmsd[i,k] - (rmsd[i,j] + rmsd[j,k])
```

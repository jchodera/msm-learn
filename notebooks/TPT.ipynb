```python
potential = lambda X : (np.sin(X[0]) + np.cos(X[1])) + np.log(sum(X**2)+1)
```

```python
x_range = np.arange(-20,20,0.1)
y_range = np.arange(-20,20,0.1)
```

```python
n = len(x_range)
```

```python
mu = np.zeros((n,n))
```

```python
for i in range(n):
    for j in range(n):
        mu[i,j] = potential(np.array([x_range[i],y_range[j]]))
```

```python
pl.imshow(mu)
pl.colorbar();
```
